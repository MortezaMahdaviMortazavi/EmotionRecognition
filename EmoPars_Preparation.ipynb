# Data Preparation

```python
# Get columns for each emotion
anger = df['Anger']
fear = df['Fear']
happiness = df['Happiness']
hatred = df['Hatred']
sadness = df['Sadness']
wonder = df['Wonder']
```

```python
# Find emotion with max value in each row
df['max_emotion'] = df[['Anger', 'Fear', 'Happiness', 'Hatred', 'Sadness', 'Wonder']].idxmax(axis=1)
```

```python
# If no emotion > 1 set as 'Other'
df['primary_emotion'] = df['max_emotion']

df.loc[df.max(axis=1) <=1, 'primary_emotion'] = 'Other'

# Map emotion labels to integers
df['primary_emotion'] = df['primary_emotion'].map({'Anger':0, 'Fear':1, 'Happiness':2,
                                                   'Hatred':3, 'Sadness':4, 'Wonder':5,
                                                   'Other':6})

print(df['primary_emotion'].value_counts())
```
